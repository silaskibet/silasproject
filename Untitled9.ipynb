```python
import pandas as pd
diab = pd.read_csv("diabetes.csv")
diab.head(12)
# check the duplicate data 
diab.duplicated().sum()
#check the shape of the  dataframe
diab.shape
#shows number of non null values and the data types
diab.info()
#DATA PREPARATION
#checking null values and there sum
diab.isnull().sum()
#diab[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']].replace(0, np.NaN)
diab.hist(figsize=(20,20),color="blue")
diab['Glucose'].fillna(diab['Glucose'].mean(), inplace = True)
diab['BloodPressure'].fillna(diab['BloodPressure'].mean(), inplace = True)
diab['SkinThickness'].fillna(diab['SkinThickness'].median(), inplace = True)
diab['Insulin'].fillna(diab['Insulin'].median(), inplace = True)
diab['BMI'].fillna(diab['BMI'].median(), inplace = True)

diab.isnull().sum()
diab.info()

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
plt.figure(figsize=(10,5))
plt.title('Diabetes Plot Yes/No', fontsize=14)
sns.countplot(diab.Outcome, data=diab, palette=('#FFFF00','#800080'))
plt.xlabel(Diabetes (0 == No, 1 == Yes)
           , fontsize=12)
plt.ylabel(Count, fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()

```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")

fig, axes = plt.subplots(2, 4, figsize=(18, 12))
fig.suptitle(
    'Diabetes Outcome Distribution WRT All Independent Variables', fontsize=16)

sns.boxplot(ax=axes[0, 0], x=diab['Outcome'], y=diab['Pregnancies'],
            hue=diab['Outcome'], palette=('#FFFF00', '#800080'))
axes[0, 0].set_title('Diabetes Outcome vs Pregnancies', fontsize=12)

sns.boxplot(ax=axes[0, 1], x=diab['Outcome'], y=diab['Glucose'],
            hue=diab['Outcome'], palette=('#FFFF00', '#800080'))
axes[0, 1].set_title('Diabetes Outcome vs Glucose', fontsize=12)

sns.boxplot(ax=axes[0, 2], x=diab['Outcome'], y=diab['BloodPressure'],
            hue=diab['Outcome'], palette=('#FFFF00', '#800080'))
axes[0, 2].set_title('Diabetes Outcome vs BloodPressure', fontsize=12)

sns.boxplot(ax=axes[0, 3], x=diab['Outcome'], y=diab['SkinThickness'],
            hue=diab['Outcome'], palette=('#FFFF00', '#800080'))
axes[0, 3].set_title('Diabetes Outcome vs SkinThickness', fontsize=12)

sns.boxplot(ax=axes[1, 0], x=diab['Outcome'], y=diab['Insulin'],
            hue=diab['Outcome'], palette=('#FFFF00', '#800080'))
axes[1, 0].set_title('Diabetes Outcome vs Insulin', fontsize=12)

sns.boxplot(ax=axes[1, 1], x=diab['Outcome'], y=diab['BMI'],
            hue=diab['Outcome'], palette=('#FFFF00', '#800080'))
axes[1, 1].set_title('Diabetes Outcome vs BMI', fontsize=12)

sns.boxplot(ax=axes[1, 2], x=diab['Outcome'], y=diab['DiabetesPedigreeFunction'],
            hue=diab['Outcome'], palette=('#FFFF00', '#800080'))
axes[1, 2].set_title(
    'Diabetes Outcome vs DiabetesPedigreeFunction', fontsize=12)

sns.boxplot(ax=axes[1, 3], x=diab['Outcome'], y=diab['Age'],
            hue=diab['Outcome'], palette=('#FFFF00', '#800080'))
axes[1, 3].set_title('Diabetes Outcome vs Age', fontsize=12)
```
